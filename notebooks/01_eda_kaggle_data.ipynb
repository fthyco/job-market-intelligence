```python
def row_quality_pipeline(df,weights=None,drop_threshold=0.85,default_weight=1,low_q=0.2,high_q=0.8):
    
    df = df.copy()
    missing_rate = df.isna().mean()

    selected_features = missing_rate[missing_rate < drop_threshold].index.tolist()

    if weights is None:
        weights = {}

    adjusted_weights = {}

    for col in selected_features:
        base_w = weights.get(col, default_weight)
        adjusted_w = base_w * (1 - missing_rate[col])
        adjusted_weights[col] = adjusted_w

    def compute_score(row):
        total_weight = 0
        score = 0
        
        for col, w in adjusted_weights.items():
            total_weight += w
            if pd.notna(row[col]):
                score += w
        
        return score / total_weight if total_weight > 0 else 0

    df["row_score"] = df.apply(compute_score, axis=1)

    low = df["row_score"].quantile(low_q)
    high = df["row_score"].quantile(high_q)

    def decision(score):
        if score < low:
            return "drop"
        elif score < high:
            return "review"
        else:
            return "keep"

    df["decision"] = df["row_score"].apply(decision)

    total_rows = len(df)
    counts = df["decision"].value_counts()

    summary = {
        "total_rows": total_rows,
        "keep_count": counts.get("keep", 0),
        "review_count": counts.get("review", 0),
        "drop_count": counts.get("drop", 0),
        "keep_pct": counts.get("keep", 0) / total_rows,
        "review_pct": counts.get("review", 0) / total_rows,
        "drop_pct": counts.get("drop", 0) / total_rows,
    }

    return df, summary
```

```python
weights = {
    "title": 10,
    "description": 10,
    "location": 9,
    "company_name": 8,
    "formatted_experience_level": 6,
    "work_type": 6,
    "min_salary": 5,
    "max_salary": 5,
    "normalized_salary": 6,
    "views": 2,
    "applies": 3,
}

df_processed, summary = row_quality_pipeline(df,weights=weights,drop_threshold=0.85)

print(summary)
```
